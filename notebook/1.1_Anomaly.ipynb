# Import Data and Check types

```python
df = pd.read_csv('../data/raw/realKnownCause/realKnownCause/ambient_temperature_system_failure.csv')
df.dtypes
df.head()
df.tail()
```

- The frequency of the data is hourly.

#### Change timestamp to datetime64 and check basic stats

```python
df.timestamp = pd.to_datetime(df.timestamp)
df.dtypes
df.head()
df.describe()
```

### Time Series Plot

```python
df.plot(x='timestamp', y= 'value', figsize = (12,8))
```

### Monthly Visualization

```python
df['year'] = df['timestamp'].apply(lambda x : x.year)
df['month'] = df['timestamp'].apply(lambda x : x.month)
df_month = df[['month', 'value']].groupby('month').count()
df_month.plot(figsize = (8,5))
```

- We can find that the aggregate temperature for the system was minimum in september and maximum in January/December. 

### Inter Quartile Range:

```python
stat = df.describe()
iqr = stat.value[6] - stat.value[4]
df['anomaly_iqr'] = [0 if (i>= (stat.value[4] - 1.5 * iqr)) & (i<= (stat.value[6] + 1.5 * iqr)) else 1 for i in df['value']]
fig, ax = plt.subplots(figsize = (12,8))
df_iqr= df.loc[df['anomaly_iqr'] == 1, ['timestamp', 'value']]
ax.plot(df['timestamp'], df['value'])
ax.scatter(df_iqr['timestamp'], df_iqr['value'], color='red')
plt.show()
```

- This simple model just captures global maxima and minima with one more local minima as outlier.

## One Class Svm:

```python
model_oc = OneClassSVM(nu = 0.05, gamma = 0.00001, kernel = 'rbf')
pred_oc = model_oc.fit_predict(df['value'].values.reshape(-1,1))
df['anomaly_oc'] = [1 if i == -1 else 0 for i in pred_oc]
```

```python
fig, ax = plt.subplots(figsize = (12,8))
df_oc= df.loc[df['anomaly_oc'] == 1, ['timestamp', 'value']]
ax.plot(df['timestamp'], df['value'])
ax.scatter(df_oc['timestamp'], df_oc['value'], color='red')
plt.show()
```

- The red dots in the plot is the detected anomaly for the dataset. Mostly the upward spike or downward spike are considered anomaly.

### Isolation Forest

```python
model_if = IsolationForest(contamination = 0.05)
model_if = model_if.fit_predict(df['value'].values.reshape(-1,1))
df['anomaly_if'] = [1 if i == -1 else 0 for i in model_if]
```

```python
fig, ax = plt.subplots(figsize = (12,8))
df_if= df.loc[df['anomaly_if'] == 1, ['timestamp', 'value']]
ax.plot(df['timestamp'], df['value'])
ax.scatter(df_if['timestamp'], df_if['value'], color='red')
plt.show()
```

- The plot looks similar in both the cases. The upper bounded and lower bould values are called anomalous by these models. 
- Let us compare the accuracy score for these models using the labels available from https://github.com/numenta/NAB/blob/master/labels/combined_windows.json .
- Add column anamoly with value 1 if the timestamp is in the range of anamolies otherwise 0.

```python
anomalies = [
    [
        "2013-12-15 07:00:00.000000",
        "2013-12-30 09:00:00.000000"
    ],
    [
        "2014-03-29 15:00:00.000000",
        "2014-04-20 22:00:00.000000"
    ]
]
df['anomaly'] = 0
for start, end in anomalies:
    df.loc[((df['timestamp'] >= start) & (df['timestamp'] <= end)), 'anomaly'] = 1
```

```python
acc_oc = accuracy_score(df['anomaly'], df['anomaly_oc'])
print("Accuracy Score for one class svm model: ", acc_oc)
```

```python
acc_if = accuracy_score(df['anomaly'], df['anomaly_if'])
print("Accuracy Score for Isolation Forest model: ", acc_if)
```

- Based on these two scores, both the model have similar score and the plot looks similar too.
- The parameters for both of the models are tuned using this accuracy.
